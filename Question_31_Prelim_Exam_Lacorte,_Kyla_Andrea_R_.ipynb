2. Write a Python script using OpenCV that:

    Reads an image named "nature.jpg"
    Converts it to grayscale
    Resizes it to 300x300 pixels
    Saves the result as "processed.jpg"

```python
import matplotlib.pyplot as plt
import cv2
```

**No. 1 Reads image named "nature.jpg"**

```python
image_path = "/content/drive/MyDrive/Colab Notebooks/nature.jpg"
img = cv2.imread(image_path)
```

**No. 2 Converts image to grayscale**

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

**No. 3 Resizes image to 300x300 pixels**

```python
resized_img = cv2.resize(gray_img, (300, 300))
```

**No. 4 Saves it as "processed.jpg"**

```python
output_path = "/content/drive/MyDrive/Colab Notebooks/processed.jpg"
cv2.imwrite(output_path, resized_img)
```
