You have an image named circles.png containing various circles  circles Links to an external site. of different sizes. Write a Python script using OpenCV that:

    Load and preprocess the image
    Detect circles using Hough Circle Transform
    Draw detected circles in red color
    Save the result as circles_detected.png

**No. 1 Load and preprocesses the image**

```python
import cv2
import numpy as np

image = cv2.imread("/content/drive/MyDrive/Colab Notebooks/circles.png")
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray = cv2.GaussianBlur(gray, (9, 9), 1.5)
```

**No. 2 Detect circles using Hough Circle Transform**

```python
circles = cv2.HoughCircles(
    gray,
    cv2.HOUGH_GRADIENT,
    dp=1,
    minDist=50,
    param1=100,
    param2=30,
    minRadius=10,
    maxRadius=100
)
```

**No. 3 Draw detected circles in red color**

```python
circles = np.uint16(np.around(circles))

for x, y, r in circles[0]:
    cv2.circle(image, (x, y), r, (0, 0, 255), 2)
    cv2.circle(image, (x, y), 2, (0, 0, 255), 3)
```

**No. 4 Save the result as circles_detected.png**

```python
cv2.imwrite("/content/drive/MyDrive/Colab Notebooks/circles_detected.png", image)
```
